```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.index = range(1,len(df)+1)
df
```

```python
df.isna()
```

```python
print(df['Duration'].dtype)
print(df['Pulse'].dtype)
print(df['Maxpulse'].dtype)
print(df['Calories'].dtype)
print(df['Date'].dtype)
```

```python
print(df['Duration'].describe())
print(df['Pulse'].describe())
print(df['Maxpulse'].describe())
print(df['Calories'].describe())
print(df['Date'].describe())
```

```python
pulse_array = df['Pulse'].to_numpy()
pulse_array
```

```python
df['Duration'].unique()
```

```python
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
```

```python
df['Calories'].mean()
```

```python
df['Calories'].fillna(df['Calories'].mean(), inplace=True)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df
```

```python
df['Month']=df['Date'].dt.month
df['Month']
```

```python
df[df['Duration']>45]
```

```python
df['Calories per minute'] = df['Calories']/df['Duration']
df
```

```python
df['Pulse'] = (df['Pulse'] - df['Pulse'].min())/(df['Pulse'].max() - df['Pulse'].min())
df
```

```python
print(df['Pulse'].describe())
print(df['Pulse'].mode())
print(df['Pulse'].median())
```

```python
df['Pulse'].corr(df['Calories'])
```

```python
df[df['Maxpulse'] == df['Maxpulse'].max()]
```

```python
df['Calories_RollingAvg'] = df['Calories'].rolling(window=5).mean()
df
```
